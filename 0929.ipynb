```python
import codecs
import csv
import prettytable
import requests
import io
import prettytable


r = requests.get("https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx",
                 params = {"DATA":"BE7EE168-4234-4E4A-91AD-616B01BEFECB"})
table=prettytable.PrettyTable(["site_id","village","marry_pair"], encodeing="utf-8")
# print(r.text)
# r.encoding="utf-8"
data = r.json()
print(table)
for i in data["responseData"]:
    print( i["site_id"],i["village"],i["marry_pair"] )
```
